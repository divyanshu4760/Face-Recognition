# FACE RECOGNITION
## Implementation of a Face Recognition system using a Pre-trained model
In this, a face recognition is run on a single image and as an additional functionality and ease of use, a box is drawn on the face of the person identified.

```python
import numpy as np
import face_recognition
from PIL import Image, ImageDraw
from IPython.display import display
```

```python
#Display an image
dis_img = Image.open('Images/Chandler_bing.jfif')
display(dis_img)
```

```python
#Helper functions

def image_load():
    """
    Loads the image database
    """
    
    face_database={}
    
    face_database["Chandler Bing"] = face_recognition.load_image_file("Images/Chandler_bing.jfif")
    face_database["Joey Tribianni"] = face_recognition.load_image_file("Images/Joey_tribianni.jfif")
    face_database["Phoebe Buffey"] = face_recognition.load_image_file("Images/Phoebe_Buffey.jfif")
    face_database["Ron Weasley"] = face_recognition.load_image_file("Images/Ron_weasley.jfif")
    
    return face_database


def img_to_encoding(faces):
    """
    Encodes the images and returns two lists: names and their encodings
    """
    
    encodings=[]
    encoding_names=[]
    
    for name, image in faces.items():
        
        encodings.append(face_recognition.face_encodings(image)[0])
        encoding_names.append(name)
        
    return encoding_names, encodings
```

#### Loading the images

```python
face_database = image_load()
```

## Learning from Examples
>*Encoding_names* are supposed to have each person's name

>*Encodings* are supposed to have 128-dimensional encoding of each person's face.

```python
encoding_names, encodings = img_to_encoding(face_database)
```

## Recognition

```python
def who_is_this(input_image, encoding_names, encodings):
    """
    Tells if face(s) in an image are from a given dataset or not
    inputs: 
    unknown_image: unknown image to be recognized
    encoding_names: names of the faces available
    encodings: Encodings of the available faces
    
    output: a PIL image contatining label on the face
    """
    
    
    # detecting all the faces and their encodings in the given image
    face_locations = face_recognition.face_locations(input_image)
    face_encodings = face_recognition.face_encodings(input_image, face_locations)

    # Converting the image to a PIL-format image so as to draw on top of it with the Pillow library
    # and Creating a Pillow ImageDraw Draw instance
    pil_img = Image.fromarray(input_image)
    draw = ImageDraw.Draw(pil_img)

    
    # Going through all the found faces in the input image
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        
        # See if the face is a match for the known face(s)
        matches = face_recognition.compare_faces(encodings, face_encoding, tolerance=0.6)

        name = "Unknown"

        # smallest Euclidean distance to the new face and index of matching face
        face_distances = face_recognition.face_distance(encodings, face_encoding)
        most_prob_ix = np.argmin(face_distances)
        
        
        if matches[most_prob_ix]:
            name = encoding_names[most_prob_ix]
            print('This person looks like "{}" with a Euclidean distance of {:.5f}'.format(name, min(face_distances)))
        else:
            print("Unknown Person!")

            
        # Draw a box around the face using the Pillow module
        draw.rectangle(((left, top), (right, bottom)), outline=(0, 255, 255))

        # Draw a label with a name below the face
        text_width, text_height = draw.textsize(name.split()[0])
        draw.rectangle(((left, bottom + text_height), (right, bottom)), fill=(0, 0, 0), outline=(255, 0, 0))
        draw.text(((left + right)/2 - text_width/2, bottom), name.split()[0], fill=(255, 255, 255, 255))
        
    del draw
    
    
    return pil_img
```
